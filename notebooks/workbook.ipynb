#### Exercise 2.3
##### Using the meteorite data from the `Meteorite_Landings.csv` file, create a pivot table that shows both the number of meteorites and the 95th percentile of meteorite mass for those that were found versus observed falling per year from 2005 through 2009 (inclusive). Hint: Be sure to convert the `year` column to a number as we did in the previous exercise.

```python
import pandas as pd
import numpy as np

# Step 1: Read the CSV and filter data for years 2005-2009
df = pd.read_csv('../data/Meteorite_Landings.csv')

df['year'] = pd.to_datetime(df['year'], format='%m/%d/%Y %I:%M:%S %p',errors='coerce').dt.year
df['year'] = df['year'].fillna(0).astype(int)

# Filter data for years 2005-2009
df_filtered = df[df['year'].between(2005, 2009)]

# Step 2: Calculate counts and 95th percentile using pivot_table
def percentile_95(x):
    return np.percentile(x, 95)

pivot_table = df_filtered.pivot_table(
    index='year',
    columns='fall',
    values='mass (g)',
    aggfunc={'mass (g)': ['count', percentile_95]}
)

pivot_table
```

#### Exercise 2.5
##### Using the taxi trip data in the `2019_Yellow_Taxi_Trip_Data.csv` file, resample the data to an hourly frequency based on the dropoff time. Calculate the total `trip_distance`, `fare_amount`, `tolls_amount`, and `tip_amount`, then find the 5 hours with the most tips.

```python
import pandas as pd
df = pd.read_csv('../data/2019_Yellow_Taxi_Trip_Data.csv', parse_dates=['tpep_dropoff_datetime'])
df.columns

# Resample to hourly frequency based on dropoff time
df.set_index('tpep_dropoff_datetime', inplace=True)
hourly_data = df.resample('h').sum()

# Sort by tip_amount and get the top 5 hours
top_hours_by_tips = hourly_data.nlargest(5, 'tip_amount')

print("Top 5 Hours with the Most Tips:")
print(top_hours_by_tips[['tip_amount']])
```
