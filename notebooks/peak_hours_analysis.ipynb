```python
import pandas as pd

data = pd.read_csv("../data/2019_Yellow_Taxi_Trip_Data.csv")
```

```python
data.rename(columns={
    'tpep_pickup_datetime':'pickup_time',
    'tpep_dropoff_datetime':'dropoff_time',
}, inplace=True)
data.head()
```

```python
# Objective: Identify peak hours for taxi trips.
# Steps:

# Group the data by hour and count the number of trips in each hour.
# Visualize the number of trips per hour using line charts or bar charts.
```

```python
data['pickup_hour'] = pd.to_datetime(data['pickup_time']).dt.hour
hourly_counts = data.groupby('pickup_hour').size().reset_index(name='count')
hourly_counts
```

```python
import matplotlib.pyplot as plt

plt.figure(figsize=(10, 6))
plt.plot(hourly_counts['pickup_hour'], hourly_counts['count'], marker='o', linestyle='-')
plt.title('Number of Trips per Hour')
plt.xlabel('Hour of the Day')
plt.ylabel('Number of Trips')
plt.xticks(hourly_counts['pickup_hour'])
plt.grid(True)
plt.show()
```
